```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_knn = pd.read_excel('result/knn.xlsx')
data_svm = pd.read_excel('result/svm.xlsx')
data_dt = pd.read_excel('result/dt.xlsx')
data_xgb = pd.read_excel('result/xgb.xlsx')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_knn, color="lightblue", showmeans=True,
            meanprops={"marker": "o", "markerfacecolor": "green", "markeredgecolor": "black"},
            medianprops={"color": "blue", "linewidth": 2},
            flierprops={"marker": "x", "color": "red", "markersize": 5})

# Customize the plot
plt.title("Box Plot of SHAP stability of KNN")
plt.grid(False)
plt.xlabel("Class Imbalance rate")
plt.ylabel("CV value")

# Show the plot
plt.show()
```

# F1 comparision plot

```python
# Data provided for F1 scores
data = {
    "Imbalance Percentage": ["1%", "5%", "10%", "20%", "30%", "40%", "50%"],
    "KNN": [0.630137, 0.75, 0.75, 0.734177215, 0.808989, 0.720930233, 0.8],
    "SVM": [0.823529, 0.809524, 0.823529412, 0.765432099, 0.776471, 0.764044944, 0.817204],
    "Decision Tree": [0.305085, 0.837209, 0.80952381, 0.891304348, 0.717949, 0.741573034, 0.76087],
    "XG-Boost": [0.823529, 0.863636, 0.850574713, 0.924731183, 0.860215, 0.842105263, 0.833333]
}
# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame to verify
print(df) 
```

```python
df.set_index("Imbalance Percentage", inplace=True)

# Plot the chart
df.plot(kind='line', marker='o', figsize=(10, 6))

# Customize the plot
plt.title("Model Performance Across Different Imbalance Percentages")
plt.xlabel("Imbalance Percentage")
plt.ylabel("F1 score")
plt.grid(True)
plt.legend(title="Model")
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
